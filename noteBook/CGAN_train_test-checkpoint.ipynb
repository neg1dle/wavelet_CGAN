```python
# ==========================================================
# Attack Integrated Script (GAN_train ç”¨ æ”»æ’ƒå®Ÿè¡Œ)
# FGSM / CW / BOTH å¯¾å¿œï¼ˆGPUå¯¾å¿œ / ä¿å­˜å‡¦ç†ä¿®æ­£æ¸ˆã¿ï¼‰
# ==========================================================

import os, re, torch, numpy as np, pandas as pd
import torchvision.transforms as T
from torchvision import models
from PIL import Image
from pathlib import Path
from tqdm import tqdm
import foolbox as fb
import torch.nn as nn
from torch.nn import functional as F


# ==========================================================
# 0) å®Ÿè¡Œãƒ¢ãƒ¼ãƒ‰
# ==========================================================
run_mode = "FGSM"     # â† FGSM / CW / both


# ==========================================================
# 1) GAN_train ã®ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆãƒ‘ã‚¹
# ==========================================================
dataset_dirs = [
    r"C:\Users\sit\wavelet_CGAN\test2\GAN_train\data1"
]


# ==========================================================
# 2) å…±é€šè¨­å®š
# ==========================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

transform_base = T.Compose([
    T.Resize((224,224)),
    T.ToTensor()
])

weights = models.MobileNet_V2_Weights.IMAGENET1K_V1
mobilenetv2 = models.mobilenet_v2(weights=weights).to(device).eval()

# â˜… transforms() ã‹ã‚‰è‡ªå‹•çš„ã«å–ã‚‹ï¼ˆKeyError å›é¿ï¼‰
tinfo = weights.transforms()
MEAN, STD = tinfo.mean, tinfo.std

imagenet_classes = weights.meta.get("categories",
                                   [f"class_{i}" for i in range(1000)])

# Normalize Module
class NormalizeModule(nn.Module):
    def __init__(self, mean, std):
        super().__init__()
        self.register_buffer("mean", torch.tensor(mean).view(1,3,1,1))
        self.register_buffer("std",  torch.tensor(std).view(1,3,1,1))
    def forward(self, x):
        return (x - self.mean) / self.std

norm_layer = NormalizeModule(MEAN, STD)
model = nn.Sequential(norm_layer, mobilenetv2).to(device).eval()

fmodel = fb.PyTorchModel(model, bounds=(0,1))

# ==========================================================
# Utility
# ==========================================================
def decode_img(path):
    img = Image.open(path).convert("RGB")
    return transform_base(img).unsqueeze(0)


# GPU/CPUå®‰å…¨TensoråŒ–
def to_torch_clipped(x):
    if isinstance(x, np.ndarray):
        t = torch.from_numpy(x)
    else:
        t = x.clone().detach()
    if t.dim() == 3:
        t = t.unsqueeze(0)
    return t.float().cpu().clamp(0,1)


def save_tensor_image(t, path):
    t = t.detach().cpu()
    arr = (t.squeeze(0).numpy().transpose(1,2,0) * 255).astype(np.uint8)
    Image.fromarray(arr).save(path)


def save_noise_vis(orig, adv, path):
    o = orig.squeeze(0).detach().cpu().numpy()
    a = adv.squeeze(0).detach().cpu().numpy()
    diff = o - a
    mag = np.max(np.abs(diff)) + 1e-8
    vis = diff / (2 * mag) + 0.5
    arr = (vis.transpose(1,2,0) * 255).astype(np.uint8)
    Image.fromarray(arr).save(path)


def extract_number(p: Path):
    m = re.search(r'\d+', p.stem)
    return int(m.group()) if m else 999999


def pred(x):
    with torch.no_grad():
        out = model(x.to(device))
        prob = F.softmax(out, 1)
        conf, idx = torch.max(prob, 1)
        idx = idx.item()
        return idx, imagenet_classes[idx], float(conf.item())


# ==========================================================
# FGSMï¼ˆå¼·ãƒã‚¤ã‚ºç‰ˆï¼šeps 0.1ã€œ1.0ï¼‰
# ==========================================================
def perform_fgsm(images, labels,
                 eps_min=0.1, eps_max=1.0, steps=1000):

    attack = fb.attacks.FGSM()

    eps_values = np.linspace(eps_min, eps_max, steps)

    orig_idx, _, _ = pred(images)

    for eps in eps_values:
        raw, clipped, is_adv = attack(fmodel, images, labels, epsilons=[eps])

        # â˜… Foolbox å‡ºåŠ›ã¯ list â†’ tensor ã‚’å–ã‚Šå‡ºã™
        if isinstance(clipped, (list, tuple)):
            clipped = clipped[0]

        # cpuå‘ã‘ã«æ•´å½¢
        clipped_t = to_torch_clipped(clipped)

        adv_idx, _, _ = pred(clipped_t)

        if adv_idx != orig_idx:
            return clipped_t, eps, adv_idx

    return None, eps_max, orig_idx

# ==========================================================
# CW
# ==========================================================
def perform_cw(images, labels, steps=100, binary=10, confidence=0.0):
    attack = fb.attacks.L2CarliniWagnerAttack(
        steps=steps,
        binary_search_steps=binary,
        confidence=confidence
    )
    raw, clipped, is_adv = attack(fmodel, images, labels, epsilons=None)

    if clipped is None:
        return None, None, None

    clipped_t = to_torch_clipped(clipped)
    clipped_gpu = clipped_t.to(device)

    l2 = torch.norm((images - clipped_gpu).reshape(1,-1), p=2).item()

    adv_idx,_,_ = pred(clipped_gpu)
    orig_idx,_,_ = pred(images)

    if adv_idx != orig_idx:
        return clipped_t, l2, adv_idx
    return None, l2, orig_idx


# ==========================================================
# â˜… æ”»æ’ƒå®Ÿè¡Œï¼ˆGAN_train ç”¨ãƒ•ãƒ«å¯¾å¿œï¼‰
# ==========================================================
def run_attack(dataset_path, attack_type):

    dataset_path = Path(dataset_path)

    print(f"\n==============================")
    print(f"â–¶ {dataset_path.name} â†’ {attack_type} Attack Start")
    print(f"==============================")

    # â˜… attackå‡ºåŠ›å…ˆï¼ˆGAN_trainä»•æ§˜ï¼‰
    output_dir = Path(str(dataset_path) + f"_{attack_type}")
    noise_dir = output_dir / "noise"
    tensor_dir = output_dir / "tensor"

    for d in [output_dir, noise_dir, tensor_dir]:
        d.mkdir(parents=True, exist_ok=True)

    AE_results = []

    image_files = sorted(
        [f for f in dataset_path.iterdir() if f.suffix.lower() in [".jpg", ".jpeg", ".png"]],
        key=extract_number
    )

    for i, img_path in enumerate(tqdm(image_files)):
        images = decode_img(img_path).to(device)
        orig_idx, orig_name, orig_conf = pred(images)
        labels = torch.tensor([orig_idx], dtype=torch.long, device=device)

        if attack_type == "FGSM":
            clipped, val, adv_lb = perform_fgsm(images, labels)
        else:
            clipped, val, adv_lb = perform_cw(images, labels)

        if clipped is not None:
            adv_idx, adv_name, adv_conf = pred(clipped.to(device))

            tensor_path = tensor_dir / f"{img_path.stem}.pt"
            torch.save(clipped.cpu(), tensor_path)

            save_tensor_image(clipped, output_dir / f"{img_path.stem}.jpg")
            save_noise_vis(images, clipped, noise_dir / f"noise_{i}.jpg")

            row = {
                "image": img_path.name,
                "Original_label": orig_idx,
                "Original_label_name": orig_name,
                "Original_confidence": orig_conf,
                "Advs_label": adv_idx,
                "Advs_label_name": adv_name,
                "Advs_confidence": adv_conf,
                "Tensor_path": str(tensor_path)
            }

            if attack_type == "FGSM":
                row["Advs_eps"] = float(val)
            else:
                row["L2_Distance"] = float(val)

            AE_results.append(row)

    df = pd.DataFrame(AE_results)
    df.to_csv(output_dir / f"{attack_type}_Results.csv", index=False, encoding="utf-8-sig")

    print(f"â–¶ å®Œäº†: {attack_type} â†’ {output_dir}")


# ==========================================================
# â˜… å…¨ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆã«å¯¾ã—ã¦å®Ÿè¡Œ
# ==========================================================
for dataset in dataset_dirs:

    if run_mode.lower() == "both":
        run_attack(dataset, "FGSM")
        run_attack(dataset, "CW")
    else:
        run_attack(dataset, run_mode)

```

```python
# ==========================================================
# Waveletåˆ†é¡å›å¾©ï¼ˆGAN_train ç”¨ å®Œå…¨å¯¾å¿œç‰ˆï¼‰
# ==========================================================

import os
import torch
import torch.nn.functional as F
import torchvision.transforms as T
from torchvision import models
from torchvision.utils import save_image
import numpy as np
import pandas as pd
import pywt
from pathlib import Path
from tqdm import tqdm


# ==========================================================
# === Waveletãƒã‚¤ã‚ºé™¤å» ===
# ==========================================================
def wavelet_denoise(img_tensor, wavelet='haar', level=1, threshold=0.07):
    img_tensor = img_tensor.detach().cpu()

    if img_tensor.ndim == 3:
        img_tensor = img_tensor.unsqueeze(0)

    img_np = img_tensor.numpy()
    denoised = []

    for b in range(img_np.shape[0]):
        out_c = []
        for c in range(img_np[b].shape[0]):
            coeffs = pywt.wavedec2(img_np[b, c], wavelet, level=level)
            cA, cD = coeffs[0], coeffs[1:]
            cD_new = []

            for (cH, cV, cD_) in cD:
                cH = pywt.threshold(cH, threshold*np.max(np.abs(cH)))
                cV = pywt.threshold(cV, threshold*np.max(np.abs(cV)))
                cD_ = pywt.threshold(cD_, threshold*np.max(np.abs(cD_)))
                cD_new.append((cH, cV, cD_))

            coeffs_new = [cA] + cD_new
            rec = pywt.waverec2(coeffs_new, wavelet)

            rec = rec[:img_np.shape[2], :img_np.shape[3]]
            out_c.append(rec)

        denoised.append(np.stack(out_c))

    return torch.tensor(np.stack(denoised)).float().clamp(0,1)


# ==========================================================
# === MobileNetV2 ãƒ­ãƒ¼ãƒ‰ ===
# ==========================================================
def load_model(device):
    model = models.mobilenet_v2(
        weights=models.MobileNet_V2_Weights.IMAGENET1K_V1
    ).to(device).eval()

    normalize = T.Normalize(
        mean=[0.485,0.456,0.406],
        std=[0.229,0.224,0.225]
    )
    return model, normalize


def predict(model, img_tensor, normalize):
    with torch.no_grad():
        preds = F.softmax(model(normalize(img_tensor)), dim=1)
        conf, idx = preds.max(1)
        return int(idx.item()), float(conf.item())


# ==========================================================
# === Waveletå‡¦ç†ãƒ¡ã‚¤ãƒ³(GAN_trainå¯¾å¿œ) ===
# ==========================================================
def process_wavelet_folder(folder):

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    folder = Path(folder)

    tensor_dir = folder / "tensor"

    if not tensor_dir.exists():
        print(f"[SKIP] tensor ãŒç„¡ã„ã®ã§ã‚¹ã‚­ãƒƒãƒ— â†’ {folder}")
        return

    print(f"\n=============================================")
    print(f"â–¶ Waveletå‡¦ç†é–‹å§‹: {folder.name}")
    print("=============================================")

    # å‡ºåŠ›ãƒ•ã‚©ãƒ«ãƒ€
    output_root = folder / "wavelet_eval"
    out_tensor_dir = output_root / "wavelet_tensor"
    out_image_dir = output_root / "wavelet_images"
    out_tensor_dir.mkdir(parents=True, exist_ok=True)
    out_image_dir.mkdir(parents=True, exist_ok=True)

    model, normalize = load_model(device)
    categories = models.MobileNet_V2_Weights.IMAGENET1K_V1.meta["categories"]

    results = []

    for fname in tqdm(sorted(os.listdir(tensor_dir))):
        if not fname.endswith(".pt"):
            continue

        stem = fname.replace(".pt", "")
        t = torch.load(tensor_dir / fname)

        if t.ndim == 3:
            t = t.unsqueeze(0)

        # Wavelet
        w = wavelet_denoise(t)

        # åˆ†é¡ç”¨GPU
        w_gpu = w.to(device)

        # äºˆæ¸¬
        label_id, conf = predict(model, w_gpu, normalize)
        label_name = categories[label_id]

        # === ä¿å­˜ ===
        torch.save(w.cpu(), out_tensor_dir / f"{stem}_wavelet.pt")
        save_image(w.cpu(), out_image_dir / f"{stem}_wavelet.png")

        results.append({
            "ImageId": stem,
            "Wavelet_Label": label_name,
            "Wavelet_Confidence": round(conf, 6),
        })

    out_csv = output_root / f"{folder.name}_Wavelet_class_results.csv"
    pd.DataFrame(results).to_csv(out_csv, index=False, encoding="utf-8-sig")

    print(f"â¡ å®Œäº†: {out_csv}")


# ==========================================================
# â˜… GAN_train ç”¨ datasets Ã— attacks è‡ªå‹•å®Ÿè¡Œ â˜…
# ==========================================================

BASE = Path(r"C:\Users\sit\wavelet_CGAN\test\GAN_train")

datasets = ["data1", "data2"]
attacks  = ["FGSM", "CW"]

for ds in datasets:
    for atk in attacks:

        target_folder = BASE / f"{ds}_{atk}"

        if target_folder.exists():
            process_wavelet_folder(target_folder)
        else:
            print(f"[SKIP] {target_folder} ã¯å­˜åœ¨ã—ã¾ã›ã‚“")
```

```python
# -*- coding: utf-8 -*-
# ==========================================================
# pix2pix (CGAN) å­¦ç¿’ã‚³ãƒ¼ãƒ‰ GAN_train å®Œå…¨å¯¾å¿œç‰ˆï¼ˆå¾©å…ƒã‚³ãƒ¼ãƒ‰ã¨Generatoræ§‹æˆã‚’çµ±ä¸€ï¼‰
# Waveletå¾ŒAEï¼ˆAï¼‰ â†’ Cleanï¼ˆBï¼‰ ã®ãƒšã‚¢å­¦ç¿’
# ==========================================================

import os
import time
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision.utils import save_image
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# ==========================================================
# === ãƒ•ã‚©ãƒ«ãƒ€è¨­å®šï¼ˆGAN_trainæ§‹æˆã«å®Œå…¨å¯¾å¿œï¼‰
# ==========================================================
BASE = Path(r"C:\Users\sit\wavelet_CGAN\test2\GAN_train")

PAIR_A = BASE / "pair_dataset" / "A" #å…¥åŠ›
PAIR_B = BASE / "pair_dataset" / "B" #ç›®æ¨™(clean)

OUT_ROOT = BASE / "pix2pix_train"
OUT_IMG = OUT_ROOT / "samples"
OUT_WGT = OUT_ROOT / "weights"
OUT_LOG = OUT_ROOT / "logs"

OUT_IMG.mkdir(parents=True, exist_ok=True)
OUT_WGT.mkdir(parents=True, exist_ok=True)
OUT_LOG.mkdir(parents=True, exist_ok=True)

# ==========================================================
# === ãƒã‚¤ãƒ‘ãƒ¼ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿
# ==========================================================
NUM_EPOCHS   = 1000
BATCH_SIZE   = 8
LEARNING_RATE = 2e-4
LAMBDA_L1    = 100
DEVICE       = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ==========================================================
# === Datasetï¼ˆA: WaveletAE, B: Cleanï¼‰
# ==========================================================
class PairDataset(Dataset):
    def __init__(self, dir_A, dir_B):
        self.dir_A = Path(dir_A)
        self.dir_B = Path(dir_B)
        self.files = sorted([f for f in os.listdir(self.dir_A) if f.endswith(".pt")])

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fname = self.files[idx]

        # A: wavelet
        a_tensor = torch.load(self.dir_A / fname)
        if a_tensor.ndim == 4:
            a_tensor = a_tensor.squeeze(0)

        # B: clean
        clean_id = fname.replace("_wavelet.pt", ".pt")
        b_tensor = torch.load(self.dir_B / clean_id)
        if b_tensor.ndim == 4:
            b_tensor = b_tensor.squeeze(0)

        # [0,1] â†’ [-1,1]
        a_tensor = (a_tensor - 0.5) / 0.5
        b_tensor = (b_tensor - 0.5) / 0.5

        return a_tensor, b_tensor


# ==========================================================
# === Generatorï¼ˆå¾©å…ƒã‚³ãƒ¼ãƒ‰ã¨åŒä¸€æ§‹æˆã® U-Net Generatorï¼‰
#     â€» test2 ã® Waveletï¼‹CGANå¾©å…ƒã‚³ãƒ¼ãƒ‰ã® UNetGenerator ã¨æƒãˆã‚‹
# ==========================================================
class UNetGenerator(nn.Module):
    def __init__(self, in_channels=3, out_channels=3, features=64):
        super().__init__()
        # encoder / decoder ã¯å¾©å…ƒã‚³ãƒ¼ãƒ‰å´ã¨åŒã˜æ§‹é€ ã«ã™ã‚‹ã“ã¨ãŒé‡è¦
        self.encoder = nn.Sequential(
            self.block(in_channels,      features,       normalize=False),
            self.block(features,         features * 2),
            self.block(features * 2,     features * 4),
            self.block(features * 4,     features * 8),
            self.block(features * 8,     features * 8),
        )

        self.decoder = nn.Sequential(
            self.upblock(features * 8,     features * 8),
            self.upblock(features * 8 * 2, features * 4),
            self.upblock(features * 4 * 2, features * 2),
            self.upblock(features * 2 * 2, features),
            nn.ConvTranspose2d(features * 2, out_channels, kernel_size=4, stride=2, padding=1),
            nn.Tanh(),  # â€» å¾©å…ƒã‚³ãƒ¼ãƒ‰ã§ã¯ decoder[-2] ã®å‡ºåŠ›ã‚’ä½¿ã†ãŒã€æ§‹é€ ã¯ã“ã®ã¾ã¾
        )

    def block(self, in_c, out_c, normalize=True):
        layers = [nn.Conv2d(in_c, out_c, 4, 2, 1)]
        if normalize:
            layers.append(nn.BatchNorm2d(out_c))
        layers.append(nn.LeakyReLU(0.2))
        return nn.Sequential(*layers)

    def upblock(self, in_c, out_c):
        return nn.Sequential(
            nn.ConvTranspose2d(in_c, out_c, 4, 2, 1),
            nn.BatchNorm2d(out_c),
            nn.ReLU()
        )

    def forward(self, x):
        skips = []
        # encoder éƒ¨åˆ†ï¼ˆå¾©å…ƒã‚³ãƒ¼ãƒ‰å´ã¨åŒã˜ï¼‰
        for layer in self.encoder:
            x = layer(x)
            skips.append(x)
        skips = skips[:-1][::-1]  # æœ€çµ‚å±¤ã‚’é™¤å¤–ã—ã¦é€†é †

        # decoder éƒ¨åˆ†ï¼šå¾©å…ƒã‚³ãƒ¼ãƒ‰ã§ã¯ decoder[:-2] ã‚’å›ã—ã¦ skip æ¥ç¶š
        for i, layer in enumerate(self.decoder[:-2]):
            x = layer(x)
            if i < len(skips):
                x = torch.cat([x, skips[i]], dim=1)

        # â˜… å¾©å…ƒã‚³ãƒ¼ãƒ‰å´ã¨åŒã˜ãã€Œdecoder[-2] ã®å‡ºåŠ›ã€ã‚’è¿”ã™
        x = self.decoder[-2](x)   # ConvTranspose2d(out_channels)
        # Tanh ã¯ decoder[-1] ã ãŒã€å¾©å…ƒã‚³ãƒ¼ãƒ‰ã® forward ã‚‚ã“ã“ã§æ­¢ã¾ã‚‹
        return x


# ==========================================================
# === PatchGAN Discriminatorï¼ˆå…ƒã®ã¾ã¾ã§OKï¼‰
# ==========================================================
class PatchDiscriminator(nn.Module):
    def __init__(self, in_c=6, features=64):
        super().__init__()
        self.model = nn.Sequential(
            self._block(in_c, features, False),
            self._block(features, features * 2),
            self._block(features * 2, features * 4),
            nn.Conv2d(features * 4, 1, 4, 1, 1),
            nn.Sigmoid()
        )

    def _block(self, in_c, out_c, norm=True):
        layers = [nn.Conv2d(in_c, out_c, 4, 2, 1)]
        if norm:
            layers.append(nn.BatchNorm2d(out_c))
        layers.append(nn.LeakyReLU(0.2))
        return nn.Sequential(*layers)

    def forward(self, a, b):
        # a: å…¥åŠ›ï¼ˆwaveletAEï¼‰, b: å‡ºåŠ›ï¼ˆcleanï¼‰
        return self.model(torch.cat([a, b], dim=1))


# ==========================================================
# === pix2pix å­¦ç¿’ãƒ«ãƒ¼ãƒ—
# ==========================================================
def train_pix2pix():
    dataset = PairDataset(PAIR_A, PAIR_B)
    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

    gen  = UNetGenerator().to(DEVICE)          # â† å¾©å…ƒã‚³ãƒ¼ãƒ‰ã¨äº’æ›ã® Generator
    disc = PatchDiscriminator().to(DEVICE)

    opt_g = torch.optim.Adam(gen.parameters(),  lr=LEARNING_RATE, betas=(0.5, 0.999))
    opt_d = torch.optim.Adam(disc.parameters(), lr=LEARNING_RATE, betas=(0.5, 0.999))

    bce = nn.BCELoss()
    l1  = nn.L1Loss()

    G_log, D_log = [], []

    print("\n==============================")
    print(" pix2pix å­¦ç¿’é–‹å§‹ï¼ˆGAN_trainãƒ»å¾©å…ƒã‚³ãƒ¼ãƒ‰äº’æ›ç‰ˆï¼‰")
    print("==============================\n")

    start = time.time()

    for epoch in range(1, NUM_EPOCHS + 1):
        loop = tqdm(loader, desc=f"Epoch {epoch}/{NUM_EPOCHS}")
        g_loss_sum, d_loss_sum = 0.0, 0.0

        for a_tensor, b_tensor in loop:
            a_tensor = a_tensor.to(DEVICE)
            b_tensor = b_tensor.to(DEVICE)

            # =======================
            # Discriminator ã®å­¦ç¿’
            # =======================
            fake = gen(a_tensor).detach()
            real_pred = disc(a_tensor, b_tensor)
            fake_pred = disc(a_tensor, fake)

            d_real = bce(real_pred, torch.ones_like(real_pred))
            d_fake = bce(fake_pred, torch.zeros_like(fake_pred))
            d_loss = (d_real + d_fake) / 2

            opt_d.zero_grad()
            d_loss.backward()
            opt_d.step()

            # =======================
            # Generator ã®å­¦ç¿’
            # =======================
            fake = gen(a_tensor)
            pred_fake = disc(a_tensor, fake)

            adv_loss = bce(pred_fake, torch.ones_like(pred_fake))
            l1_loss_val = l1(fake, b_tensor) * LAMBDA_L1
            g_loss = adv_loss + l1_loss_val

            opt_g.zero_grad()
            g_loss.backward()
            opt_g.step()

            g_loss_sum += g_loss.item()
            d_loss_sum += d_loss.item()

        G_log.append(g_loss_sum / len(loader))
        D_log.append(d_loss_sum / len(loader))

        # --- ã‚µãƒ³ãƒ—ãƒ«ç”»åƒä¿å­˜ï¼ˆ[-1,1] â†’ [0,1] ã«æˆ»ã—ã¦ä¿å­˜ï¼‰ ---
        with torch.no_grad():
            fake_vis = (fake * 0.5 + 0.5).clamp(0, 1)
            save_image(fake_vis, OUT_IMG / f"sample_epoch{epoch}.png")

        # --- é‡ã¿ä¿å­˜ ---
        # 1. æœ€çµ‚ã‚¨ãƒãƒƒã‚¯ã§å¿…ãšä¿å­˜
        if epoch == NUM_EPOCHS:
            save_flag = True
        # 2. 100ã‚¨ãƒãƒƒã‚¯ä»¥ä¸‹ã®å ´åˆã€10ã‚¨ãƒãƒƒã‚¯ã”ã¨ã«ä¿å­˜
        elif epoch <= 100 and epoch % 10 == 0:
            save_flag = True
        # 3. 100ã‚¨ãƒãƒƒã‚¯ã‚ˆã‚Šå¤§ãã„å ´åˆã€100ã‚¨ãƒãƒƒã‚¯ã”ã¨ã«ä¿å­˜
        elif epoch > 100 and epoch % 100 == 0:
            save_flag = True
        # 4. ãã‚Œä»¥å¤–ã¯ä¿å­˜ã—ãªã„
        else:
            save_flag = False

        if save_flag:
            torch.save(gen.state_dict(), OUT_WGT / f"gen_weights_epoch{epoch}.pth")
            print(f"\nâœ” Generator weights saved to: {OUT_WGT / f'gen_weights_epoch{epoch}.pth'}")

    # --- Lossãƒ­ã‚°ä¿å­˜ ---
    df = pd.DataFrame({
        "Epoch": range(1, NUM_EPOCHS + 1),
        "G_loss": G_log,
        "D_loss": D_log
    })
    df.to_csv(OUT_LOG / "loss_history.csv", index=False)

    plt.figure(figsize=(10, 5))
    plt.plot(G_log, label="Generator Loss")
    plt.plot(D_log, label="Discriminator Loss")
    plt.legend()
    plt.grid()
    plt.savefig(OUT_LOG / "loss_curve.png")
    plt.close()

    print("\n=== å­¦ç¿’å®Œäº† ===")
    print(f"å‡ºåŠ›ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒª: {OUT_ROOT}")
    print(f"å­¦ç¿’æ™‚é–“: {(time.time() - start) / 60:.2f} min")
    print("=======================================")

# ==========================================================
# å®Ÿè¡Œ
# ==========================================================
if __name__ == "__main__":
    train_pix2pix()
```

```python
# -*- coding: utf-8 -*-
# ==========================================================
# pix2pix (CGAN) å­¦ç¿’ã‚³ãƒ¼ãƒ‰ GAN_train å®Œå…¨å¯¾å¿œç‰ˆï¼ˆå¾©å…ƒã‚³ãƒ¼ãƒ‰ã¨Generatoræ§‹æˆã‚’çµ±ä¸€ï¼‰
# Waveletå¾ŒAEï¼ˆAï¼‰ â†’ Cleanï¼ˆBï¼‰ ã®ãƒšã‚¢å­¦ç¿’
# ==========================================================

import os
import time
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from torchvision.utils import save_image
from tqdm import tqdm
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# ==========================================================
# === ãƒ•ã‚©ãƒ«ãƒ€è¨­å®šï¼ˆGAN_trainæ§‹æˆã«å®Œå…¨å¯¾å¿œï¼‰
# ==========================================================
BASE = Path(r"C:\Users\sit\wavelet_CGAN\test2\GAN_train")

PAIR_A = BASE / "pair_dataset" / "A"
PAIR_B = BASE / "pair_dataset" / "B"

OUT_ROOT = BASE / "pix2pix_train_normal"
OUT_IMG = OUT_ROOT / "samples"
OUT_WGT = OUT_ROOT / "weights"
OUT_LOG = OUT_ROOT / "logs"

OUT_IMG.mkdir(parents=True, exist_ok=True)
OUT_WGT.mkdir(parents=True, exist_ok=True)
OUT_LOG.mkdir(parents=True, exist_ok=True)

# ==========================================================
# === ãƒã‚¤ãƒ‘ãƒ¼ãƒ‘ãƒ©ãƒ¡ãƒ¼ã‚¿
# ==========================================================
NUM_EPOCHS   = 1000
BATCH_SIZE   = 8
LEARNING_RATE = 2e-4
LAMBDA_L1    = 100
DEVICE       = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ==========================================================
# === Datasetï¼ˆA: WaveletAE, B: Cleanï¼‰
# ==========================================================
class PairDataset(Dataset):
    def __init__(self, dir_A, dir_B):
        self.dir_A = Path(dir_A)
        self.dir_B = Path(dir_B)
        self.files = sorted([f for f in os.listdir(self.dir_A) if f.endswith(".pt")])

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fname = self.files[idx]

        # A: wavelet
        a_tensor = torch.load(self.dir_A / fname)
        if a_tensor.ndim == 4:
            a_tensor = a_tensor.squeeze(0)

        # B: clean
        clean_id = fname.replace("_wavelet.pt", ".pt")
        b_tensor = torch.load(self.dir_B / clean_id)
        if b_tensor.ndim == 4:
            b_tensor = b_tensor.squeeze(0)

        # [0,1] â†’ [-1,1]
        a_tensor = (a_tensor - 0.5) / 0.5
        b_tensor = (b_tensor - 0.5) / 0.5

        return a_tensor, b_tensor


# ==========================================================
# === Generatorï¼ˆå¾©å…ƒã‚³ãƒ¼ãƒ‰ã¨åŒä¸€æ§‹æˆã® U-Net Generatorï¼‰
#     â€» test2 ã® Waveletï¼‹CGANå¾©å…ƒã‚³ãƒ¼ãƒ‰ã® UNetGenerator ã¨æƒãˆã‚‹
# ==========================================================
class UNetGenerator(nn.Module):
    def __init__(self, in_channels=3, out_channels=3, features=64):
        super().__init__()
        # encoder / decoder ã¯å¾©å…ƒã‚³ãƒ¼ãƒ‰å´ã¨åŒã˜æ§‹é€ ã«ã™ã‚‹ã“ã¨ãŒé‡è¦
        self.encoder = nn.Sequential(
            self.block(in_channels,      features,       normalize=False),
            self.block(features,         features * 2),
            self.block(features * 2,     features * 4),
            self.block(features * 4,     features * 8),
            self.block(features * 8,     features * 8),
        )

        self.decoder = nn.Sequential(
            self.upblock(features * 8,     features * 8),
            self.upblock(features * 8 * 2, features * 4),
            self.upblock(features * 4 * 2, features * 2),
            self.upblock(features * 2 * 2, features),
            nn.ConvTranspose2d(features * 2, out_channels, kernel_size=4, stride=2, padding=1),
            nn.Tanh(),  # â€» å¾©å…ƒã‚³ãƒ¼ãƒ‰ã§ã¯ decoder[-2] ã®å‡ºåŠ›ã‚’ä½¿ã†ãŒã€æ§‹é€ ã¯ã“ã®ã¾ã¾
        )

    def block(self, in_c, out_c, normalize=True):
        layers = [nn.Conv2d(in_c, out_c, 4, 2, 1)]
        if normalize:
            layers.append(nn.BatchNorm2d(out_c))
        layers.append(nn.LeakyReLU(0.2))
        return nn.Sequential(*layers)

    def upblock(self, in_c, out_c):
        return nn.Sequential(
            nn.ConvTranspose2d(in_c, out_c, 4, 2, 1),
            nn.BatchNorm2d(out_c),
            nn.ReLU()
        )

    def forward(self, x):
        skips = []
        # encoder éƒ¨åˆ†ï¼ˆå¾©å…ƒã‚³ãƒ¼ãƒ‰å´ã¨åŒã˜ï¼‰
        for layer in self.encoder:
            x = layer(x)
            skips.append(x)
        skips = skips[:-1][::-1]  # æœ€çµ‚å±¤ã‚’é™¤å¤–ã—ã¦é€†é †

        # decoder éƒ¨åˆ†ï¼šå¾©å…ƒã‚³ãƒ¼ãƒ‰ã§ã¯ decoder[:-2] ã‚’å›ã—ã¦ skip æ¥ç¶š
        for i, layer in enumerate(self.decoder[:-2]):
            x = layer(x)
            if i < len(skips):
                x = torch.cat([x, skips[i]], dim=1)

        # â˜… å¾©å…ƒã‚³ãƒ¼ãƒ‰å´ã¨åŒã˜ãã€Œdecoder[-2] ã®å‡ºåŠ›ã€ã‚’è¿”ã™
        x = self.decoder[-2](x)   # ConvTranspose2d(out_channels)
        # Tanh ã¯ decoder[-1] ã ãŒã€å¾©å…ƒã‚³ãƒ¼ãƒ‰ã® forward ã‚‚ã“ã“ã§æ­¢ã¾ã‚‹
        return x


# ==========================================================
# === PatchGAN Discriminatorï¼ˆå…ƒã®ã¾ã¾ã§OKï¼‰
# ==========================================================
class PatchDiscriminator(nn.Module):
    def __init__(self, in_c=6, features=64):
        super().__init__()
        self.model = nn.Sequential(
            self._block(in_c, features, False),
            self._block(features, features * 2),
            self._block(features * 2, features * 4),
            nn.Conv2d(features * 4, 1, 4, 1, 1),
            nn.Sigmoid()
        )

    def _block(self, in_c, out_c, norm=True):
        layers = [nn.Conv2d(in_c, out_c, 4, 2, 1)]
        if norm:
            layers.append(nn.BatchNorm2d(out_c))
        layers.append(nn.LeakyReLU(0.2))
        return nn.Sequential(*layers)

    def forward(self, a, b):
        # a: å…¥åŠ›ï¼ˆwaveletAEï¼‰, b: å‡ºåŠ›ï¼ˆcleanï¼‰
        return self.model(torch.cat([a, b], dim=1))


# ==========================================================
# === pix2pix å­¦ç¿’ãƒ«ãƒ¼ãƒ—
# ==========================================================
def train_pix2pix():
    dataset = PairDataset(PAIR_A, PAIR_B)
    loader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

    gen  = UNetGenerator().to(DEVICE)          # â† å¾©å…ƒã‚³ãƒ¼ãƒ‰ã¨äº’æ›ã® Generator
    disc = PatchDiscriminator().to(DEVICE)

    opt_g = torch.optim.Adam(gen.parameters(),  lr=LEARNING_RATE, betas=(0.5, 0.999))
    opt_d = torch.optim.Adam(disc.parameters(), lr=LEARNING_RATE, betas=(0.5, 0.999))

    bce = nn.BCELoss()
    l1  = nn.L1Loss()

    G_log, D_log = [], []

    print("\n==============================")
    print(" pix2pix å­¦ç¿’é–‹å§‹ï¼ˆGAN_trainãƒ»å¾©å…ƒã‚³ãƒ¼ãƒ‰äº’æ›ç‰ˆï¼‰")
    print("==============================\n")

    start = time.time()

    for epoch in range(1, NUM_EPOCHS + 1):
        loop = tqdm(loader, desc=f"Epoch {epoch}/{NUM_EPOCHS}")
        g_loss_sum, d_loss_sum = 0.0, 0.0

        for a_tensor, b_tensor in loop:
            a_tensor = a_tensor.to(DEVICE)
            b_tensor = b_tensor.to(DEVICE)

            # =======================
            # Discriminator ã®å­¦ç¿’
            # =======================
            fake = gen(a_tensor).detach()
            real_pred = disc(a_tensor, b_tensor)
            fake_pred = disc(a_tensor, fake)

            d_real = bce(real_pred, torch.ones_like(real_pred))
            d_fake = bce(fake_pred, torch.zeros_like(fake_pred))
            d_loss = (d_real + d_fake) / 2

            opt_d.zero_grad()
            d_loss.backward()
            opt_d.step()

            # =======================
            # Generator ã®å­¦ç¿’
            # =======================
            fake = gen(a_tensor)
            pred_fake = disc(a_tensor, fake)

            adv_loss = bce(pred_fake, torch.ones_like(pred_fake))
            l1_loss_val = l1(fake, b_tensor) * LAMBDA_L1
            g_loss = adv_loss + l1_loss_val

            opt_g.zero_grad()
            g_loss.backward()
            opt_g.step()

            g_loss_sum += g_loss.item()
            d_loss_sum += d_loss.item()

        G_log.append(g_loss_sum / len(loader))
        D_log.append(d_loss_sum / len(loader))

        # --- ã‚µãƒ³ãƒ—ãƒ«ç”»åƒä¿å­˜ï¼ˆ[-1,1] â†’ [0,1] ã«æˆ»ã—ã¦ä¿å­˜ï¼‰ ---
        with torch.no_grad():
            fake_vis = (fake * 0.5 + 0.5).clamp(0, 1)
            save_image(fake_vis, OUT_IMG / f"sample_epoch{epoch}.png")

        # --- é‡ã¿ä¿å­˜ ---
        # 1. æœ€çµ‚ã‚¨ãƒãƒƒã‚¯ã§å¿…ãšä¿å­˜
        if epoch == NUM_EPOCHS:
            save_flag = True
        # 2. 100ã‚¨ãƒãƒƒã‚¯ä»¥ä¸‹ã®å ´åˆã€10ã‚¨ãƒãƒƒã‚¯ã”ã¨ã«ä¿å­˜
        elif epoch <= 100 and epoch % 10 == 0:
            save_flag = True
        # 3. 100ã‚¨ãƒãƒƒã‚¯ã‚ˆã‚Šå¤§ãã„å ´åˆã€100ã‚¨ãƒãƒƒã‚¯ã”ã¨ã«ä¿å­˜
        elif epoch > 100 and epoch % 100 == 0:
            save_flag = True
        # 4. ãã‚Œä»¥å¤–ã¯ä¿å­˜ã—ãªã„
        else:
            save_flag = False

        if save_flag:
            torch.save(gen.state_dict(), OUT_WGT / f"gen_weights_epoch{epoch}.pth")
            print(f"\nâœ” Generator weights saved to: {OUT_WGT / f'gen_weights_epoch{epoch}.pth'}")

    # --- Lossãƒ­ã‚°ä¿å­˜ ---
    df = pd.DataFrame({
        "Epoch": range(1, NUM_EPOCHS + 1),
        "G_loss": G_log,
        "D_loss": D_log
    })
    df.to_csv(OUT_LOG / "loss_history.csv", index=False)

    plt.figure(figsize=(10, 5))
    plt.plot(G_log, label="Generator Loss")
    plt.plot(D_log, label="Discriminator Loss")
    plt.legend()
    plt.grid()
    plt.savefig(OUT_LOG / "loss_curve.png")
    plt.close()

    print("\n=== å­¦ç¿’å®Œäº† ===")
    print(f"å‡ºåŠ›ãƒ‡ã‚£ãƒ¬ã‚¯ãƒˆãƒª: {OUT_ROOT}")
    print(f"å­¦ç¿’æ™‚é–“: {(time.time() - start) / 60:.2f} min")
    print("=======================================")

# ==========================================================
# å®Ÿè¡Œ
# ==========================================================
if __name__ == "__main__":
    train_pix2pix()
```

```python
# -*- coding: utf-8 -*-
# ==========================================================
# CGANå˜ç‹¬å¾©å…ƒï¼‹3æ®µéšæ¯”è¼ƒï¼†æˆåŠŸç‡åˆ¤å®š (Waveletã‚¹ã‚­ãƒƒãƒ—)
# ==========================================================

import os
import sys
import torch
import torch.nn.functional as F
import pandas as pd
from pathlib import Path
from torchvision import models
from torchvision.utils import save_image
from torch import nn
from tqdm import tqdm
from shutil import copy2

# ========= 0) å†ç¾æ€§ï¼ˆä»»æ„ï¼‰ =========
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

# ========= 1) å…±é€šè¨­å®š =========
BASE = Path(r"C:\Users\sit\wavelet_CGAN\test2")

DATASETS = ["data1","data2"]
ATTACKS = ["FGSM","CW"] # U+00A0ã‚’å‰Šé™¤
LEGACY_PREPROCESS = True

# CGAN Generatorã®é‡ã¿ãƒ‘ã‚¹
WEIGHT_PATH = BASE / "GAN_train" /"pix2pix_train_normal" / "weights" / "gen_weights_epoch100.pth"

# ========= 2) å­˜åœ¨ãƒã‚§ãƒƒã‚¯ =========
def _must_exist(p: Path, hint: str = ""):
    if not p.exists():
        print(f"\n[ERROR] Not found: {p}")
        if hint:
            print("Hint:", hint)
        sys.exit(1)

# ========= 3) MobileNetV2 =========
def load_classifier(device):
    weights = models.MobileNet_V2_Weights.IMAGENET1K_V1
    model = models.mobilenet_v2(weights=weights).to(device).eval()
    categories = weights.meta["categories"]

    if LEGACY_PREPROCESS:
        preprocess = weights.transforms()    # Resize/Crop å«ã‚€
    else:
        from torchvision import transforms as T
        preprocess = T.Normalize(mean=weights.meta["mean"], std=weights.meta["std"])

    return model, preprocess, categories

# ========= 4) CGAN Generatorï¼ˆæ—§ä»•æ§˜ï¼‰=========
class UNetGenerator(nn.Module):
    def __init__(self, in_channels=3, out_channels=3, features=64):
        super().__init__()
        self.encoder = nn.Sequential(
            self.block(in_channels, features, normalize=False),
            self.block(features, features * 2),
            self.block(features * 2, features * 4),
            self.block(features * 4, features * 8),
            self.block(features * 8, features * 8),
        )

        self.decoder = nn.ModuleList([
            self.upblock(features * 8, features * 8),
            self.upblock(features * 8 * 2, features * 4),
            self.upblock(features * 4 * 2, features * 2),
            self.upblock(features * 2 * 2, features),
            nn.ConvTranspose2d(features * 2, out_channels, 4, 2, 1),
            nn.Tanh(),
        ])

    def block(self, in_c, out_c, normalize=True):
        layers = [nn.Conv2d(in_c, out_c, 4, 2, 1)]
        if normalize:
            layers.append(nn.BatchNorm2d(out_c))
        layers.append(nn.LeakyReLU(0.2))
        return nn.Sequential(*layers)

    def upblock(self, in_c, out_c):
        return nn.Sequential(
            nn.ConvTranspose2d(in_c, out_c, 4, 2, 1),
            nn.BatchNorm2d(out_c),
            nn.ReLU()
        )

    def forward(self, x):
        skips = []
        for layer in self.encoder:
            x = layer(x)
            skips.append(x)

        x = skips[-1]
        skips = skips[:-1][::-1]

        for i in range(4):
            x = self.decoder[i](x)
            x = torch.cat([x, skips[i]], dim=1)

        x = self.decoder[4](x)
        return self.decoder[5](x)


# ========= 5) ãƒ‡ãƒ¼ã‚¿ã‚»ãƒƒãƒˆÃ—æ”»æ’ƒã”ã¨ã«å‡¦ç† (CGANå˜ç‹¬) =========
def process_dataset_attack(dataset: str, attack_type: str):

    print("\n" + "=" * 60)
    print(f"â–¶ é–‹å§‹: {dataset} / {attack_type} / CGANå˜ç‹¬çŸ¯æ­£")
    print("=" * 60 + "\n")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # ----- ãƒ‘ã‚¹ -----
    clean_dir = BASE / dataset
    adv_dir    = BASE / f"{dataset}_{attack_type}"

    # ã€å…¥åŠ›ãƒ†ãƒ³ã‚½ãƒ«ãƒ‘ã‚¹ã€‘æ”»æ’ƒå¾Œã®ã‚ªãƒªã‚¸ãƒŠãƒ«ãƒ†ãƒ³ã‚½ãƒ«ã‚’ç›´æ¥å‚ç…§
    tensor_dir  = adv_dir / "tensor" # â˜… æ”»æ’ƒå¾Œã®ãƒ†ãƒ³ã‚½ãƒ«ãƒ‘ã‚¹ (ä¾‹: data1_FGSM/tensor)

    # å‡ºåŠ›ãƒ•ã‚©ãƒ«ãƒ€ (åç§°å¤‰æ›´)
    cgan_out_dir = BASE / "pix2pix_restore_CGAN_ONLY" / f"{dataset}_{attack_type}"
    cgan_out_dir.mkdir(parents=True, exist_ok=True)

    # CGAN Tensor ä¿å­˜ãƒ•ã‚©ãƒ«ãƒ€
    tensor_out_dir = cgan_out_dir / "tensor_out"
    tensor_out_dir.mkdir(parents=True, exist_ok=True)

    cgan_csv    = cgan_out_dir / f"pix2pix_class_results_{dataset}_{attack_type}_CGAN_Only.csv"
    merged_csv = cgan_out_dir / f"compare_{dataset}_{attack_type}_CGAN_Only.csv"

    # æ¯”è¼ƒç”¨CSV
    before_csv  = clean_dir / "clean_dataset_labels.csv"
    after_csv   = adv_dir / f"{attack_type}_Results.csv"

    # å¤±æ•—ç”»åƒä¿å­˜å…ˆ
    cgan_failed_dir     = BASE / "cgan_restore_failed_CGAN_ONLY"     / f"{dataset}_{attack_type}"
    cgan_failed_dir.mkdir(parents=True, exist_ok=True)

    failed_list_csv = cgan_out_dir / f"failed_lists_{dataset}_{attack_type}_CGAN_Only.csv"

    # ãƒã‚§ãƒƒã‚¯
    _must_exist(before_csv)
    _must_exist(after_csv)
    _must_exist(tensor_dir, hint="æ”»æ’ƒå¾Œã®ç”»åƒãƒ†ãƒ³ã‚½ãƒ«ï¼ˆ.ptï¼‰ãŒã“ã®ãƒ‘ã‚¹ã«å­˜åœ¨ã™ã‚‹ã“ã¨ã‚’ç¢ºèªã—ã¦ãã ã•ã„ã€‚")
    _must_exist(WEIGHT_PATH)

    # ========= 5-1) CGANå¾©å…ƒï¼‹åˆ†é¡ =========
    def restore_and_classify():

        gen = UNetGenerator().to(device)
        gen.load_state_dict(torch.load(WEIGHT_PATH, map_location=device))
        gen.eval()

        clf, preprocess, categories = load_classifier(device)
        results = []

        print(f"\n=== CGANæ¨è«–ï¼‹åˆ†é¡ ({dataset}/{attack_type}) ===\n")

        with torch.no_grad():
            for fname in tqdm(sorted(os.listdir(tensor_dir))):
                if not fname.endswith(".pt"):
                    continue

                t = torch.load(tensor_dir / fname, map_location=device)
                if t.ndim == 3:
                    t = t.unsqueeze(0)

                # æ­£è¦åŒ–
                if t.min() >= 0 and t.max() <= 1:
                    t_in = (t - 0.5) / 0.5
                else:
                    t_in = t

                restored = gen(t_in.to(device))
                restored_01 = (restored * 0.5 + 0.5).clamp(0, 1)

                # ãƒ™ãƒ¼ã‚¹åã¯ Wavelet ã‚µãƒ•ã‚£ãƒƒã‚¯ã‚¹ç„¡ã—ã‚’æƒ³å®š
                base = Path(fname).stem 

                # CGAN å‡ºåŠ› Tensor ä¿å­˜
                torch.save(restored.cpu(), tensor_out_dir / f"{base}.pt")

                # PNG ä¿å­˜
                out_png = cgan_out_dir / f"{base}_{attack_type}_restored_CGAN_Only.png"
                save_image(restored_01.cpu(), out_png)

                # åˆ†é¡
                x = restored_01.squeeze(0).cpu()
                x = preprocess(x).unsqueeze(0).to(device)
                preds = F.softmax(clf(x), dim=1)

                conf, idx = preds.max(1)
                label = categories[int(idx)]

                results.append({
                    "ImageId": base,
                    "CGAN_Label": label,
                    "CGAN_Confidence": float(conf),
                })

        pd.DataFrame(results).to_csv(cgan_csv, index=False, encoding="utf-8-sig")
        print(f"âœ” CGANåˆ†é¡çµæœCSV: {cgan_csv}")
        print(f"âœ” Tensorä¿å­˜å…ˆ: {tensor_out_dir}")


    # ========= 5-2) 3æ®µéšCSVçµ±åˆ =========
    def merge_and_evaluate():

        print(f"\n=== 3æ®µéšCSVçµ±åˆ ({dataset}/{attack_type}) ===\n")

        df_b = pd.read_csv(before_csv) # (1) å…ƒç”»åƒ (Original)
        df_a = pd.read_csv(after_csv)  # (2) æ”»æ’ƒå¾Œ
        df_g = pd.read_csv(cgan_csv)    # (3) CGANå¾©å…ƒå¾Œ

        df_b["ImageId"] = df_b["image"].apply(lambda x: Path(str(x)).stem)
        df_a["ImageId"] = df_a["image"].apply(lambda x: Path(str(x)).stem)
        df_g["ImageId"] = df_g["ImageId"].astype(str)

        df_b = df_b.rename(columns={"label_name":"label_name_before",
                                     "confidence":"confidence_before"})
        df_a = df_a.rename(columns={"Advs_label_name":"label_name_after",
                                     "Advs_confidence":"confidence_after"})

        # çµ±åˆ
        df = (df_b.merge(df_a, on="ImageId", how="inner")
                  .merge(df_g, on="ImageId", how="left"))

        df["AttackSuccess"] = df["label_name_before"] != df["label_name_after"]
        df["RecoverySuccess_CGAN"]     = df["label_name_before"] == df["CGAN_Label"]

        df.to_csv(merged_csv, index=False, encoding="utf-8-sig")
        print(f"âœ” çµ±åˆCSV: {merged_csv}")

        # æˆåŠŸç‡
        total = len(df)
        atk = int(df["AttackSuccess"].sum())
        g_ok = int(df["RecoverySuccess_CGAN"].sum())
        pct = lambda n: (n/total*100.0) if total else 0.0

        print("\n=== çµ±è¨ˆ ===")
        print(f"ç·æ•°: {total}")
        print(f"æ”»æ’ƒæˆåŠŸç‡: {atk}/{total} ({pct(atk):.2f}%)")
        print(f"CGANçŸ¯æ­£æˆåŠŸç‡: {g_ok}/{total} ({pct(g_ok):.2f}%)")

        # ---- å¤±æ•—ç”»åƒä¿å­˜ (CGANã®ã¿) ----
        cgan_failed = df[df["RecoverySuccess_CGAN"] == False]
        for _, row in cgan_failed.iterrows():
            img_id = row["ImageId"]
            src = cgan_out_dir / f"{img_id}_{attack_type}_restored_CGAN_Only.png"
            if src.exists():
                copy2(src, cgan_failed_dir / src.name)

        if len(cgan_failed) > 0:
            cgan_failed.to_csv(failed_list_csv, index=False, encoding="utf-8-sig")
            print(f"ğŸ“ CGANå¤±æ•—ä¸€è¦§CSV: {failed_list_csv}")

    # ===== å®Ÿè¡Œ =====
    restore_and_classify()
    merge_and_evaluate()


# ========= 6) ä¸€æ‹¬å®Ÿè¡Œ =========
if __name__ == "__main__":

    _must_exist(WEIGHT_PATH)

    for ds in DATASETS:
        for atk in ATTACKS:
            process_dataset_attack(ds, atk)

    print("\n=== CGANå˜ç‹¬çŸ¯æ­£ å…¨å‡¦ç†å®Œäº† ===\n")
```
