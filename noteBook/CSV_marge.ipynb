```python
import pandas as pd

# --- CSVを読み込み ---
df_orig = pd.read_csv(r"C:\Users\sit\wavelet_CGAN\testdata3\original_pred.csv")
df_adv = pd.read_csv(r"C:\Users\sit\wavelet_CGAN\testdata3\data3_FGSM\Origin_Adv_simple.csv")

# --- ファイル名でマージ ---
merged = pd.merge(df_orig, df_adv, left_on="filename", right_on="image", how="inner")

# --- 現在の列名を確認 ---
print("列名一覧:", merged.columns.tolist())

# --- 列名に応じて再整形（自動対応）---
# df_adv の列が "Original_class", "Advs_class", "SucEps" 形式かどうか確認して選択
if "Advs_class" in merged.columns:
    merged = merged[[
        "filename",
        "label_name", "confidence",
        "Advs_class", "Advs_conf", "SucEps"
    ]]
    merged.columns = ["filename", "Original_label", "Original_conf", "Advs_label", "Advs_conf", "SucEps"]

else:
    # 列名が異なる場合のフォールバック
    print("⚠️ 想定外の列名です。マージ結果を確認してください。")
    print(merged.head())

# --- 保存 ---
save_path = r"C:\Users\sit\wavelet_CGAN\testdata3\data3_FGSM\Merged_Result.csv"
merged.to_csv(save_path, index=False, encoding="utf-8-sig")

print("✅ マージ結果を保存しました:", save_path)
```
